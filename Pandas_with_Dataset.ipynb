```python
# ============================================
# 1. SETUP AND LOAD DATASET
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Mount Google Drive (if using Colab)
# from google.colab import drive
# drive.mount('/content/drive')

# Load the dataset
# Option 1: If you've uploaded the file to Colab
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('Early-childhood-education-database-Mar-2025.csv', encoding='latin1')

# Option 2: If you have the file path
# df = pd.read_csv('/content/Early-childhood-education-database-Mar-2025.csv')

# ============================================
# 2. EXPLORE DATA STRUCTURE
# ============================================

print("="*80)
print("INITIAL DATA EXPLORATION")
print("="*80)

# Check shape
print(f"\nDataset shape: {df.shape}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Display first few rows
print("\nFirst 5 rows:")
df.head()

# Column names
print("\nColumn names:")
print(df.columns.tolist())

# Data types and info
print("\nDataset info:")
df.info()

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# ============================================
# 3. CLEAN THE DATA
# ============================================

print("="*80)
print("DATA CLEANING")
print("="*80)

# The dataset has a complex structure with metadata rows at the top
# Let's find where the actual data starts

# Method 1: Skip rows when reading (cleaner approach)
# Let's reload the file and skip the metadata rows
# The data starts at row where 'Afghanistan' appears

# Find the index where countries start
df_raw = pd.read_csv('Early-childhood-education-database-Mar-2025.csv', header=None, encoding='latin1')
start_idx = df_raw[df_raw[0] == 'Afghanistan'].index[0]

# Reload with proper headers
# The actual column headers are in rows 0-3, but they're complex
# Let's create clean column names

df = pd.read_csv('Early-childhood-education-database-Mar-2025.csv',
                 skiprows=start_idx,
                 names=['country', 'attendance_total', 'flag1', 'attendance_male',
                        'flag2', 'attendance_female', 'flag3', 'attendance_poorest',
                        'flag4', 'attendance_richest', 'flag5', 'source'], encoding='latin1')

print("Dataset loaded with clean column names")
print(f"Shape: {df.shape}")
df.head()

# Remove summary rows at the bottom
# Keep only country-level data, remove regional aggregates
countries_to_remove = ['SUMMARY', 'East Asia and Pacific', 'Europe and Central Asia',
                       'Eastern Europe and Central Asia', 'Western Europe',
                       'Latin America and Caribbean', 'Middle East and North Africa',
                       'North America', 'South Asia', 'Sub-Saharan Africa',
                       'Eastern and Southern Africa', 'West and Central Africa',
                       'Least developed countries', 'World', 'Notes:', 'Indicator definition:',
                       'Source:', 'Prepared by', 'Contact us:']

df = df[~df['country'].isin(countries_to_remove)]
df = df[df['country'].notna()]  # Remove any remaining NaN rows

print(f"After removing summary rows: {df.shape}")

# Convert attendance columns to numeric, coercing errors to NaN
attendance_cols = ['attendance_total', 'attendance_male', 'attendance_female',
                   'attendance_poorest', 'attendance_richest']

for col in attendance_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Handle missing values
print("\nMissing values before cleaning:")
print(df[attendance_cols].isnull().sum())

# Option 1: Drop rows where total attendance is missing (most important column)
df = df.dropna(subset=['attendance_total'], thresh=1)
print(f"\nAfter dropping rows with no total attendance: {df.shape}")

# Option 2: Fill missing values with appropriate values
# For regional analysis, we might fill with median
median_total = df['attendance_total'].median()
df['attendance_total'].fillna(median_total, inplace=True)

# Fill other attendance columns with their respective medians
for col in attendance_cols[1:]:
    df[col].fillna(df[col].median(), inplace=True)

print("\nMissing values after cleaning:")
print(df[attendance_cols].isnull().sum())

# Clean the source column - extract survey type
df['survey_type'] = df['source'].str.extract(r'(MICS|DHS|HIES|NFHS|CASEN|ENSANUT|ENDES|NHMS|SUSENAS|Eurostat|PNAD)')
df['survey_year'] = df['source'].str.extract(r'(\d{4})')
df['survey_year'] = pd.to_numeric(df['survey_year'], errors='coerce')

# Create region groups based on UNICEF regional classification
def assign_region(country):
    sub_saharan = ['Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 'Cameroon',
                   'Cabo Verde', 'Central African Republic', 'Chad', 'Comoros', 'Congo',
                   "Côte d'Ivoire", 'Democratic Republic of the Congo', 'Eswatini',
                   'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 'Guinea-Bissau',
                   'Kenya', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi', 'Mali',
                   'Mauritania', 'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda',
                   'Senegal', 'Sierra Leone', 'South Africa', 'South Sudan', 'Sudan',
                   'Togo', 'Uganda', 'United Republic of Tanzania', 'Zambia', 'Zimbabwe']

    south_asia = ['Afghanistan', 'Bangladesh', 'Bhutan', 'India', 'Iran (Islamic Republic of)',
                  'Maldives', 'Nepal', 'Pakistan', 'Sri Lanka']

    mena = ['Algeria', 'Bahrain', 'Djibouti', 'Egypt', 'Iraq', 'Jordan', 'Kuwait',
            'Lebanon', 'Libya', 'Morocco', 'Oman', 'Qatar', 'Saudi Arabia',
            'State of Palestine', 'Syrian Arab Republic', 'Tunisia', 'United Arab Emirates',
            'Yemen']

    lac = ['Argentina', 'Barbados', 'Belize', 'Bolivia (Plurinational State of)', 'Brazil',
           'Chile', 'Colombia', 'Costa Rica', 'Cuba', 'Dominica', 'Dominican Republic',
           'Ecuador', 'El Salvador', 'Grenada', 'Guatemala', 'Guyana', 'Haiti', 'Honduras',
           'Jamaica', 'Mexico', 'Nicaragua', 'Panama', 'Paraguay', 'Peru', 'Saint Lucia',
           'Saint Vincent and the Grenadines', 'Suriname', 'Trinidad and Tobago', 'Uruguay',
           'Venezuela (Bolivarian Republic of)']

    eca = ['Albania', 'Armenia', 'Azerbaijan', 'Belarus', 'Bosnia and Herzegovina',
           'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Estonia', 'Georgia', 'Hungary',
           'Kazakhstan', 'Kyrgyzstan', 'Latvia', 'Lithuania', 'Montenegro',
           'North Macedonia', 'Poland', 'Republic of Moldova', 'Romania',
           'Russian Federation', 'Serbia', 'Slovakia', 'Slovenia', 'Tajikistan',
           'Türkiye', 'Turkmenistan', 'Ukraine', 'Uzbekistan']

    eap = ['Australia', 'Brunei Darussalam', 'Cambodia', 'China', 'Cook Islands',
           'Democratic People\'s Republic of Korea', 'Fiji', 'Indonesia', 'Japan',
           'Kiribati', 'Lao People\'s Democratic Republic', 'Malaysia', 'Marshall Islands',
           'Micronesia (Federated States of)', 'Mongolia', 'Myanmar', 'Nauru', 'New Zealand',
           'Niue', 'Palau', 'Papua New Guinea', 'Philippines', 'Republic of Korea',
           'Samoa', 'Singapore', 'Solomon Islands', 'Thailand', 'Timor-Leste', 'Tokelau',
           'Tonga', 'Tuvalu', 'Vanuatu', 'Viet Nam']

    if country in sub_saharan:
        return 'Sub-Saharan Africa'
    elif country in south_asia:
        return 'South Asia'
    elif country in mena:
        return 'Middle East & North Africa'
    elif country in lac:
        return 'Latin America & Caribbean'
    elif country in eca:
        return 'Europe & Central Asia'
    elif country in eap:
        return 'East Asia & Pacific'
    else:
        return 'Other'

df['region'] = df['country'].apply(assign_region)

print("\nRegion distribution:")
print(df['region'].value_counts())

# Calculate wealth gap
df['wealth_gap'] = df['attendance_richest'] - df['attendance_poorest']

# Calculate gender gap
df['gender_gap'] = df['attendance_female'] - df['attendance_male']

# ============================================
# 4. SUMMARY STATISTICS
# ============================================

print("="*80)
print("SUMMARY STATISTICS")
print("="*80)

# Overall summary statistics
print("\nOverall Attendance Statistics:")
print(df[attendance_cols].describe())

# Individual statistics
print("\nGlobal Early Childhood Education Attendance:")
print(f"Mean attendance: {df['attendance_total'].mean():.1f}%")
print(f"Median attendance: {df['attendance_total'].median():.1f}%")
print(f"Minimum attendance: {df['attendance_total'].min():.1f}%")
print(f"Maximum attendance: {df['attendance_total'].max():.1f}%")
print(f"Standard deviation: {df['attendance_total'].std():.1f}%")

# Count of countries with data
print(f"\nCountries with data: {df['country'].nunique()}")
print(f"Countries with wealth quintile data: {df['attendance_poorest'].notna().sum()}")

# Gender statistics
print(f"\nMean male attendance: {df['attendance_male'].mean():.1f}%")
print(f"Mean female attendance: {df['attendance_female'].mean():.1f}%")
print(f"Mean gender gap (female-male): {df['gender_gap'].mean():.1f}%")

# Wealth inequality statistics
print(f"\nMean poorest quintile attendance: {df['attendance_poorest'].mean():.1f}%")
print(f"Mean richest quintile attendance: {df['attendance_richest'].mean():.1f}%")
print(f"Mean wealth gap (richest-poorest): {df['wealth_gap'].mean():.1f}%")

# Top 10 countries with highest attendance
print("\nTop 10 Countries with Highest ECE Attendance:")
top10 = df.nlargest(10, 'attendance_total')[['country', 'attendance_total', 'source']]
print(top10.to_string(index=False))

# Bottom 10 countries with lowest attendance
print("\nBottom 10 Countries with Lowest ECE Attendance:")
bottom10 = df.nsmallest(10, 'attendance_total')[['country', 'attendance_total', 'source']]
print(bottom10.to_string(index=False))

# ============================================
# 5. ANSWER 3 SPECIFIC QUESTIONS
# ============================================

print("="*80)
print("ANSWERING 3 SPECIFIC QUESTIONS")
print("="*80)

# QUESTION 1: How does early childhood education attendance vary by region?
print("\n" + "-"*60)
print("QUESTION 1: Regional analysis of ECE attendance")
print("-"*60)

# Group by region and calculate statistics
region_stats = df.groupby('region').agg({
    'attendance_total': ['mean', 'median', 'min', 'max', 'count'],
    'wealth_gap': 'mean',
    'gender_gap': 'mean'
}).round(1)

region_stats.columns = ['mean_%', 'median_%', 'min_%', 'max_%', 'country_count', 'avg_wealth_gap', 'avg_gender_gap']
region_stats = region_stats.sort_values('mean_%', ascending=False)

print("\nRegional Early Childhood Education Attendance:")
print(region_stats)

# Find regional best and worst performers
for region in df['region'].unique():
    if region != 'Other':
        region_df = df[df['region'] == region]
        if len(region_df) > 0:
            best = region_df.loc[region_df['attendance_total'].idxmax()]
            worst = region_df.loc[region_df['attendance_total'].idxmin()]
            print(f"\n{region}:")
            print(f"  Best: {best['country']} ({best['attendance_total']:.1f}%)")
            print(f"  Worst: {worst['country']} ({worst['attendance_total']:.1f}%)")

# QUESTION 2: What is the relationship between wealth and early childhood education attendance?
print("\n" + "-"*60)
print("QUESTION 2: Wealth inequality in ECE attendance")
print("-"*60)

# Calculate correlation between wealth and attendance
correlation = df['attendance_poorest'].corr(df['attendance_richest'])
print(f"Correlation between poorest and richest attendance: {correlation:.3f}")

# Countries with largest wealth gap
print("\nTop 10 countries with largest wealth gap (richest - poorest):")
largest_gap = df.nlargest(10, 'wealth_gap')[['country', 'region', 'attendance_poorest',
                                              'attendance_richest', 'wealth_gap']]
print(largest_gap.to_string(index=False))

# Countries with smallest wealth gap (most equitable)
print("\nTop 10 countries with smallest wealth gap (most equitable):")
smallest_gap = df.nsmallest(10, 'wealth_gap')[['country', 'region', 'attendance_poorest',
                                                'attendance_richest', 'wealth_gap']]
print(smallest_gap.to_string(index=False))

# Countries where poorest have higher attendance than richest (rare)
equitable = df[df['attendance_poorest'] > df['attendance_richest']]
if len(equitable) > 0:
    print(f"\nCountries where poorest attend more than richest ({len(equitable)} countries):")
    print(equitable[['country', 'attendance_poorest', 'attendance_richest', 'wealth_gap']].to_string(index=False))

# QUESTION 3: Is there a gender gap in early childhood education attendance, and does it vary by region?
print("\n" + "-"*60)
print("QUESTION 3: Gender gap analysis in ECE attendance")
print("-"*60)

# Overall gender statistics
gender_regions = df.groupby('region').agg({
    'attendance_male': 'mean',
    'attendance_female': 'mean',
    'gender_gap': 'mean'
}).round(1)

gender_regions.columns = ['male_%', 'female_%', 'gender_gap_%']
gender_regions = gender_regions.sort_values('gender_gap_%', ascending=False)

print("\nGender gap by region:")
print(gender_regions)

# Countries with largest gender gap favoring females
print("\nTop 10 countries where girls attend more than boys:")
female_favor = df.nlargest(10, 'gender_gap')[['country', 'region', 'attendance_male',
                                               'attendance_female', 'gender_gap']]
print(female_favor.to_string(index=False))

# Countries with largest gender gap favoring males
print("\nTop 10 countries where boys attend more than girls:")
male_favor = df.nsmallest(10, 'gender_gap')[['country', 'region', 'attendance_male',
                                              'attendance_female', 'gender_gap']]
print(male_favor.to_string(index=False))

# Statistical test - is gender gap significant?
from scipy import stats
male_mean = df['attendance_male'].mean()
female_mean = df['attendance_female'].mean()
t_stat, p_value = stats.ttest_rel(df['attendance_male'].dropna(),
                                  df['attendance_female'].dropna())

print(f"\nGender gap statistical significance:")
print(f"  Male mean: {male_mean:.2f}%")
print(f"  Female mean: {female_mean:.2f}%")
print(f"  Difference: {female_mean - male_mean:.2f}%")
print(f"  T-statistic: {t_stat:.3f}")
print(f"  P-value: {p_value:.4f}")
if p_value < 0.05:
    print("  Conclusion: Gender gap is statistically significant")
else:
    print("  Conclusion: Gender gap is NOT statistically significant")

# ============================================
# 6. VISUALIZATIONS (Bonus)
# ============================================

print("="*80)
print("DATA VISUALIZATIONS")
print("="*80)

# Set style
plt.style.use('ggplot')
sns.set_palette("Set2")

# Create figure with subplots
fig = plt.figure(figsize=(20, 15))

# 1. Regional comparison
ax1 = plt.subplot(3, 2, 1)
region_means = df.groupby('region')['attendance_total'].mean().sort_values()
colors = plt.cm.RdYlGn(region_means.values/100)
region_means.plot(kind='barh', ax=ax1, color=colors)
ax1.set_title('Average ECE Attendance by Region', fontsize=14, fontweight='bold')
ax1.set_xlabel('Attendance (%)')
ax1.axvline(x=df['attendance_total'].mean(), color='black', linestyle='--', label=f'Global Mean: {df["attendance_total"].mean():.1f}%')
ax1.legend()

# 2. Wealth gap distribution
ax2 = plt.subplot(3, 2, 2)
df['wealth_gap'].dropna().hist(bins=20, edgecolor='black', ax=ax2)
ax2.set_title('Distribution of Wealth Gaps (Richest - Poorest)', fontsize=14, fontweight='bold')
ax2.set_xlabel('Percentage Point Gap')
ax2.set_ylabel('Number of Countries')
ax2.axvline(x=df['wealth_gap'].mean(), color='red', linestyle='--', label=f'Mean Gap: {df["wealth_gap"].mean():.1f}%')
ax2.legend()

# 3. Gender gap comparison
ax3 = plt.subplot(3, 2, 3)
gender_data = df[['attendance_male', 'attendance_female']].melt(var_name='Gender', value_name='Attendance')
gender_data['Gender'] = gender_data['Gender'].replace({'attendance_male': 'Male', 'attendance_female': 'Female'})
sns.boxplot(x='Gender', y='Attendance', data=gender_data, ax=ax3)
ax3.set_title('Gender Comparison in ECE Attendance', fontsize=14, fontweight='bold')
ax3.set_ylabel('Attendance (%)')

# 4. Top and bottom performers
ax4 = plt.subplot(3, 2, 4)
top_bottom = pd.concat([df.nlargest(5, 'attendance_total'),
                        df.nsmallest(5, 'attendance_total')])
colors_top = ['green']*5 + ['red']*5
top_bottom.sort_values('attendance_total', ascending=True).plot(
    kind='barh', y='attendance_total', x='country', ax=ax4,
    color=colors_top, legend=False
)
ax4.set_title('Top 5 and Bottom 5 Countries', fontsize=14, fontweight='bold')
ax4.set_xlabel('Attendance (%)')

# 5. Wealth quintile comparison
ax5 = plt.subplot(3, 2, 5)
wealth_data = df[['attendance_poorest', 'attendance_richest']].melt(var_name='Quintile', value_name='Attendance')
wealth_data['Quintile'] = wealth_data['Quintile'].replace({
    'attendance_poorest': 'Poorest 20%',
    'attendance_richest': 'Richest 20%'
})
sns.boxplot(x='Quintile', y='Attendance', data=wealth_data, ax=ax5)
ax5.set_title('Wealth Quintile Comparison', fontsize=14, fontweight='bold')
ax5.set_ylabel('Attendance (%)')

# 6. Scatter plot: Poorest vs Richest
ax6 = plt.subplot(3, 2, 6)
scatter = ax6.scatter(df['attendance_poorest'], df['attendance_richest'],
                      c=df['wealth_gap'], cmap='RdYlGn_r', s=50, alpha=0.7)
ax6.set_xlabel('Poorest 20% Attendance (%)')
ax6.set_ylabel('Richest 20% Attendance (%)')
ax6.set_title('Poorest vs Richest Attendance', fontsize=14, fontweight='bold')
ax6.axline([0, 0], [1, 1], color='gray', linestyle='--', alpha=0.5)
plt.colorbar(scatter, ax=ax6, label='Wealth Gap (pp)')

plt.tight_layout()
plt.savefig('ece_education_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print("\nVisualizations saved as 'ece_education_analysis.png'")

# ============================================
# 7. SUMMARY AND CONCLUSIONS
# ============================================

print("="*80)
print("KEY FINDINGS AND CONCLUSIONS")
print("="*80)

print(f"""
KEY FINDINGS FROM UNICEF EARLY CHILDHOOD EDUCATION DATASET:

1. GLOBAL OVERVIEW:
   - Global average ECE attendance: {df['attendance_total'].mean():.1f}%
   - Range: {df['attendance_total'].min():.1f}% to {df['attendance_total'].max():.1f}%
   - Total countries analyzed: {df['country'].nunique()}

2. REGIONAL DISPARITIES:
   - Highest performing region: {region_stats.index[0]} ({region_stats.iloc[0]['mean_%']:.1f}%)
   - Lowest performing region: {region_stats.index[-1]} ({region_stats.iloc[-1]['mean_%']:.1f}%)
   - Gap between best and worst regions: {region_stats.iloc[0]['mean_%'] - region_stats.iloc[-1]['mean_%']:.1f}%

3. WEALTH INEQUALITY:
   - Average wealth gap: {df['wealth_gap'].mean():.1f} percentage points
   - Poorest quintile: {df['attendance_poorest'].mean():.1f}%
   - Richest quintile: {df['attendance_richest'].mean():.1f}%
   - Countries where richest have >50pp advantage: {(df['wealth_gap'] > 50).sum()}

4. GENDER EQUALITY:
   - Average male attendance: {df['attendance_male'].mean():.1f}%
   - Average female attendance: {df['attendance_female'].mean():.1f}%
   - Average gender gap: {df['gender_gap'].mean():.1f}% (favoring females)
   - Statistically significant: {'Yes' if p_value < 0.05 else 'No'}

5. KEY TAKEAWAYS:
   - Large disparities exist between wealthy and poor children
   - Regional differences are substantial
   - Gender gaps are relatively small compared to wealth gaps
   - Many low-income countries need targeted interventions
""")

# ============================================
# 8. EXPORT CLEANED DATASET
# ============================================

# Save cleaned dataset
df.to_csv('cleaned_ece_education_data.csv', index=False)
print("\nCleaned dataset saved as 'cleaned_ece_education_data.csv'")

# Export summary statistics
summary_stats = df[attendance_cols + ['wealth_gap', 'gender_gap']].describe()
summary_stats.to_csv('ece_education_summary_stats.csv')
print("Summary statistics saved as 'ece_education_summary_stats.csv'")
```
